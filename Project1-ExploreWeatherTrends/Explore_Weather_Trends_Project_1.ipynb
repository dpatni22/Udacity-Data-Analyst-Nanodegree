Analyzing the data and visualizing it properly by using 'Moving Average';

### Importing necessary libraries:

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

### Importing the data:

```python
df_global=pd.read_csv('GLOBAL_DATA.csv')
df_city=pd.read_csv('san_jose_data.csv')
df_global.head(10)
df_city.head(10)
```

### Merging data frames (df_global and df_city) using year column

```python
df_merged=pd.merge(df_global, df_city, on="year")
df_merged.head(15)
```

### Plotting df_global and df_city with year on X Axis and avg_temp_x, avg_temp_y on Y axis

```python
plt.plot(df_merged['year'],df_merged['avg_temp_x'],label='Global')
plt.plot(df_merged['year'],df_merged['avg_temp_y'],label='San Jose')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)") 
plt.title("Global vs San Jose Average Temperature Plot")
plt.show()
```

### Calculating Moving Averages

```python
global_mv_avg = df_merged['avg_temp_x'].rolling(10).mean()
san_jose_mv_avg = df_merged['avg_temp_y'].rolling(10).mean()
```

```python
plt.plot(df_merged['year'],df_merged['10_Year_MA_x'], label = 'Global')
plt.plot(df_merged['year'],df_merged['10_Year_MA_y'], label = 'San Jose')
plt.legend()
plt.title("Global vs San Jose Moving Average Temperature")
plt.xlabel('Year')
plt.ylabel('Moving Average Temp.')
plt.show()
```
